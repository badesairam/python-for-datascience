# MATPLOTLIB

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# can use notebook for interactive plot instead of static png
%matplotlib inline 
plt.style.use('classic')
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-')
ax.plot(x, np.cos(x), '--')
# fig.savefig('figure.png')
from IPython.display import Image
#Image('figure.png')
fig,ax = plt.subplots(nrows=2,ncols=3)
ax[1][1].plot(x,np.cos(x),':')
ax[0][0].plot(x,x**2,'-.r')
ax[0][1].plot(x, x + 1, '--c')
```

```python
x = np.linspace(0, 20, 100)
plt.title("powers")
plt.xlabel("x")
plt.ylabel("power(x)")
plt.plot(x,x,':r',label="straight")
plt.plot(x,x**2,'--g',label="square")
plt.legend()
```
